### Discount Applied column set

```python
df["Discount Applied"].fillna(df["Discount Applied"].mode()[0],inplace=True)
```

### Item column set

```python
df["Item"].fillna(df["Item"].mode()[0],inplace=True)
```

### Price Per Unit column set

```python
df["Price Per Unit"].fillna(df["Price Per Unit"].mean(),inplace=True)
round(df["Price Per Unit"], 2)
```

### Quantity column set

```python
df["Quantity"].fillna(df["Quantity"].mean(),inplace=True)
```

```python
df["Quantity"]=df["Quantity"].astype(int)
```

### Total Spent column set

```python
df["Total Spent"].fillna(df["Total Spent"].mean(),inplace=True)
```

### Customer ID sequences set 

```python
df.sort_values("Customer ID",ascending=True,inplace=True)
```
